```python
import pandas as pd

# Load CSV
df = pd.read_csv("forex_dataset.csv")

# Remove the first two bad rows
df = df.iloc[2:].copy()

# Rename first column
df.columns = ["Date", "GOLD", "EURUSD", "GBPUSD", "USDJPY"]

# Convert numeric columns
for col in ["GOLD", "EURUSD", "GBPUSD", "USDJPY"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Remove any remaining bad rows
df = df.dropna()

print(df.head())
print(df.dtypes)

# Save cleaned dataset
df.to_csv("forex_dataset_clean.csv", index=False)

print("\nвњ… Clean dataset saved as forex_dataset_clean.csv")
```
