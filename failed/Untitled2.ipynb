```python
import streamlit as st
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
from tensorflow.keras.models import load_model
```

```python
st.set_page_config(page_title="Forex AI Dashboard", layout="wide")

st.title("📊 Forex & Gold AI Signal System")
st.write("LSTM-based market prediction + signal generator")
```

```python
pair = st.selectbox(
    "Select Trading Pair",
    ["XAUUSD (Gold)", "EURUSD", "GBPUSD", "USDJPY"]
)
```

```python
@st.cache_resource
def load_selected_model(pair):
    if pair == "XAUUSD (Gold)":
        return load_model("gold_model.h5", compile = False)
    elif pair == "EURUSD":
        return load_model("eurusd_model.h5")
    elif pair == "GBPUSD":
        return load_model("gbpusd_model.h5")
    elif pair == "USDJPY":
        return load_model("usdjpy_model.h5")
```

```python
def predict_price(model, data):
    data = np.array(data)
    data = data.reshape(1, data.shape[0], 1)
    prediction = model.predict(data)
    return prediction[0][0]
```

```python
st.subheader("Input Market Data")

price_input = st.number_input("Enter Latest Close Price", value=2000.0)
```

```python
model = load_selected_model(pair)

if st.button("Predict"):
    input_data = [price_input] * 30  # dummy sequence for now
    prediction = predict_price(model, input_data)
    st.session_state["prediction"] = float(prediction)

    st.success(f"Predicted Price: {prediction:.2f}")
```

```python
if "prediction" in st.session_state:
    prediction = st.session_state["prediction"]
    if prediction > price_input:
        st.markdown("### 📈 BULLISH SIGNAL")
    else:
        st.markdown("### 📉 BEARISH SIGNAL")
else:
    st.info("Press Predict to evaluate the market signal.")
```

```python
st.warning(
    "⚠️ Market conditions change rapidly. This is not financial advice."
)
```
