```python
import yfinance as yf
import pandas as pd

pairs = {
    "USDCHF": "USDCHF=X",
    "EURUSD": "EURUSD=X",
    "GBPUSD": "GBPUSD=X",
    "USDJPY": "USDJPY=X"
}

all_data = {}

for name, symbol in pairs.items():

    print(f"Downloading {name}...")

    df = yf.download(symbol, period="5y", interval="1d")

    # IMPORTANT CHECK
    if df.empty:
        print(f"❌ No data for {name}")
        continue

    all_data[name] = df["Close"]

# 🔥 FIX: combine properly with alignment
dataset = pd.concat(all_data, axis=1)

# Drop rows with missing values
dataset = dataset.dropna()

print("\n✅ Final Dataset Preview:")
print(dataset.head())

dataset.to_csv("forex_dataset.csv")
print("\nSaved successfully!")
```
